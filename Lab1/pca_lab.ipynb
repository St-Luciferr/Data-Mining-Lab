```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
iris=datasets.load_iris()
R=np.array(iris.data)
print(R.shape)

df_iris=pd.DataFrame(iris.data,columns=iris.feature_names)
df_iris['target']=pd.Series(iris.target)
df_iris.head(10)
```

```python
iris.target_names
```

```python
X=df_iris.iloc[:,0:4].values
print(X)
print(np.mean(X))
print(np.std(X))
```

```python
X_std=StandardScaler().fit_transform(X)
print(np.mean(X_std))
print(np.std(X_std))
```

```python
norm=np.random.rand(4,4)
print(norm)
print(X_std.shape)
```

```python
covar=np.cov(X_std.T)
```

```python
covar
```

```python
eug,vec=np.linalg.eig(covar)
print(eug)
print(vec)
```

```python
prop0=eug[0]*100/np.sum(eug)
prop1=eug[1]*100/np.sum(eug)
prop2=eug[2]*100/np.sum(eug)
prop3=eug[3]*100/np.sum(eug)
print(prop0)
print(prop1)
print(prop2)
print(prop3)
```

```python
eu_vec=vec[:,0:2]
print(eu_vec.shape)
print(eu_vec)

```

```python
Sy_iris=np.matmul(X_std,eu_vec)
print(Sy_iris.shape)
```

```python
plt.scatter(Sy_iris[:,0],Sy_iris[:,1])
plt.grid()
```
